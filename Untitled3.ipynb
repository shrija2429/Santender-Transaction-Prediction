```
import folium
```

```
dir(folium)
```

```
import pandas as pd
data=pd.read_csv("Volcanoes.txt")
data
```

```
lat = list(data["LAT"])
lon = list(data["LON"])
elev = list(data["ELEV"])
```

```
map = folium.Map(location=[38.2,-99.09],zoom_start=6)
map
```

```
fg= folium.FeatureGroup(name="My Map")
for lt , ln, el in zip(lat,lon,elev):
    fg.add_child(folium.Marker(location= (lt ,ln, el) ,popup=folium.Popup(str(el)+"m",parse_html=True),icon=folium.Icon(color='green')))
map.add_child(fg)
```

```
map.save("Map1.html")
```
